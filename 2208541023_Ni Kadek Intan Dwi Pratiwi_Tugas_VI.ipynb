iloc (Index Location) digunakan untuk menyeleksi data pada lokasi tertentu saja. Cuma bedanya, jika loc menyeleksi data dari baris satu ke baris lain dan dari kolom satu ke kolom lain, maka iloc menyeleksi data dari baris satuhingga satu baris sebelum baris lain dan antara kolom satu dengan satu kolom sebelum kolom lain. Konsepnya persis seperti index pada array. Konsep iloc tidak mengenal nama kolom. Nama kolom akan diganti dengan index kolom yang dimulai dari nol hingga seterusnya.

```python
# memanggil bebarapa data pada semua kolom dengan menggunakan fungsi iloc
# ketik print (variabel.iloc[baris pada data yang ingin dipanggil])
# diperhatikan tanda titik pada penulisan kodenya
print(bunga_iris.iloc[3])
# sehingga hasil yang muncul adalah data pada baris 3 pada semua kolom
```

```python
# memanggil bebarapa data pada semua jenis kolom menggunakan interval
# ketik print (variabel[interval yang ingin dipanggil datanya])
# pada interval nya baris baris pertama dan kedua dipisah dengan tanda :, seperti contoh dibawah ini
print(bunga_iris[0:8])
# pada kode diatas memanggil pada data 0:8
# yang aartinya memanggil data pada baris ke 0 sampai baris sebelum 8 yaitu baris 7
```

### 5.21.3 Mengurutkan data pada Pandas

Data yang telah dipanggil dapat diurutkan mulai dari tinggi ke rendah (atau
dari rendah ke tinggi) untuk data berupa angka, sedangkan data berupa string
akan diurutkan berdasarkan urutan abjad.

```python
# mengurutkan data dari yang rendah ke paling tinggi
# menggunakan sort_values 

# ketik variabelnya.sort_values("masukan data yang ingin diurutkan")
# diperhatikan penggunaan tanda titik sebagai pemisah antara variabel dengan fungsi sort_values
# diperhatikan penggunaan "" atau '' pada data yang ingin dipanggil
bunga_iris.sort_values("sepal length (cm)")
# sehingga hasil yang muncul semua data, tetapi data pada sepal length (cm) yang diurutkan dari kecil ke tinggi
```

```python
# mengurutkan data dari yang tinggi ke rendah
# menggunakan fungsi sort_values, dibarengi dengan penggunaan ascending = False
# sort_values artinya mengurutkan data dari yang terendah ke terbesar
# penggunaan ascending = False artinya membalik urutan pengurutan
# Berarti urutan terbesar ke terkecil

# ketik variabelnya.sort_values("masukan data yang ingin diurutkan", ascending = False)
# diperhatikan penggunaan tanda titik sebagai pemisah antara variabel dengan sort_values
# diperhatikan penggunaan "" atau '' pada data yang ingin dipanggil
bunga_iris.sort_values("sepal length (cm)", ascending = False)
# karena ascending = False artinya membalik urutan pengurutan, hasil yang muncul adalah kebalikan pada contoh diatas
```

```python
# mengurutkan dua data dari yang rendah ke paling tinggi
# menggunakan sort_values 

# ketik variabelnya.sort_values("masukan data yang ingin diurutkan 1","masukan data yang ingin diurutkan 2")
# diperhatikan penggunaan tanda titik sebagai pemisah antara variabel dengan fungsi sort_values
# diperhatikan penggunaan "" atau '' pada data yang ingin dipanggil
bunga_iris.sort_values(["sepal length (cm)","sepal width (cm)"])
# pada contoh ini data yang prioritaskan untuk diurutkan adalah data 1, apabila data 1 jumlahnya sama, bari mengurutkan ke data 2
# maka tidak heran apabila pada data 2 tidak terlihat berurutan dari rendah ke tinggi
```

### 5.21.4 Mengubah data pada Pandas

Data yang telah dipanggil dapat dirubah sesuai dengan kebutuhan.

```python
# pada contoh ini mencoba untuk mengubah data yang dipanggil
# disini mencoba untuk mengalikan 10 pada sepal length (cm)

# ketik variabel["masukan data yang ingin dirubah"] = 10*vriabel["data yang ingin dirubah"]
# diperhatikan penggunaan tanda []
# diperhatikan juga penggunaan '' atau "", seperti contoh dibawah ini
bunga_iris["sepal length (cm)"] = 10*bunga_iris["sepal length (cm)"] 
# print variabelnya untuk menampilkan data yang sudah dirubah
print (bunga_iris)
# hasil yang muncul adalah data pada sepal length dikali 10
```

```python
# mengembalikan data seperti semula 
# agar kode dibawahnya tidak menggunakan data yang sudah dirubah
# contoh diatas mengalikan dengan 10, untuk ini dikali dengan 1/10
# kode mirip dengan contoh diatas 
bunga_iris["sepal length (cm)"] = 1/10*bunga_iris["sepal length (cm)"] 
print (bunga_iris)
```

```python
# menghilangkan data dengan menggunakan drop

# ketik variabel = variabel.drop(columns = ["data pada kolom apa yang ingin dihilangkan"])
# terdapat tanda titik untuk pemisah variabel dan penggunaan drop
# diperhatikan penggunaan tanda "" atau '' pada data kolom yang ingin dihilangkan
bunga_iris = bunga_iris.drop(columns=["sepal width (cm)"])
print (bunga_iris)
# sehingga hasil yang muncul data pada sepal width (cm) terhapus
```

```python
# disini untuk mengembalikan data 
# dengan kembali membaca file menggunakan link
# dengan cara yang sama pada contoh membaca file dengan menggunakan link
bunga_iris = pd.read_csv("https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/bunga_iris.csv")
print(bunga_iris)
# hasil yang muncul adalah data seperti semula tanpa perubahan
```

```python
# menambahkan data baru pada kolom 
# disini akan menambahkan luas sepal, yang dimna perkalian antara sepal length dengan sepal width

# ketik variabel ["nama data baru"] = variabel["data 1"] * variabel["data 2"]
# tanda * artinya perkalian 
# diperhatikan tanda "" atau '' saat memanggil data
# diperhatikan juga tanda []
bunga_iris["Luas Sepal"] = bunga_iris["sepal length (cm)"] *  bunga_iris["sepal width (cm)"]
print (bunga_iris)
# maka hasil yang muncul adalah bertambahnya data yaitu luas sepal
# dengan data diperoleh dari perkalian sepal length dan sepal width
```

### 5.21.6 Menyaring data menggunakan Pandas

Data yang telah dipanggil dapat disaring sesuai dengan kebutuhan analisis kedepannya.

```python
# menyaring data pada nilai tertentu 
# menggunakan loc
# loc berfungsi sebagai loc menyeleksi data dari baris satu ke baris lain dan dari kolom satu ke kolom lain

# masukan variabel.loc[variabel["data yang ingin disaring"] == nilai data yang ingin disaring]
# diperhatikan penggunaan = dua kali
# terdapat tanda titik sebagai pemisah variabel dengan loc
# diperhatikan penggunaan "" pada memanggil data
bunga_iris.loc[bunga_iris["petal length (cm)"] == 1.4]
# karena memanggil data pada petal length (cm) dengan nilai 1.4
# maka hasil yang muncul adalah petal length dengan nilai 1.4 dengan data lainnya menyesuaikan
```

```python
# menyaring data pada dua data dengan nilai tertentu 
# menggunakan loc
# loc berfungsi sebagai loc menyeleksi data dari baris satu ke baris lain dan dari kolom satu ke kolom lain

# masukan variabel.loc[(variabel["data 1"] == nilai data 1) & (variabel["data 2"] == nilai data 2) ]
# diperhatikan penggunaan = dua kali
# terdapat tanda titik sebagai pemisah variabel dengan loc
# diperhatikan penggunaan "" pada memanggil data
# digunakan tanda & sebagai tanda dan
bunga_iris.loc[(bunga_iris["petal length (cm)"] == 1.4) &
              (bunga_iris["sepal length (cm)"] == 5.1 )]
# pada contoh ini menyaring dua data
# memanggil data pada petal length (cm) = 1.4 dan sepal length (cm) = 5.1
# sehingga dari banyaknya data hanya ada dua data yang dapat disaring
```

### 5.21.7 Pengelompokan data menggunakan Pandas

Data yang sama dapat dikelompokkan untuk analisis lebih lanjut.

```python
# menambahkan data nama bunga untuk lebih gampang dalam menegurutkan data
# menambahkan data dengan menggunakan link 
# caranya sama dengan contoh sebelumnya 
# masukan variabel["nama data baru"] = pd.read_csv ('link data')
# diperhatikan tanda "" atau ""
bunga_iris["flower name"] = pd.read_csv ("https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/nama_bunga.csv")
print (bunga_iris)
# sehingga hasil yang muncul adalah data bunga iris ditambah dengan flower_name
```

```python
# mengelompokkan data yang sama dengan menggunakan groupby 
# mengelompokan data dengan nama bunga yang sama untuk mencari nilai rata rata di setiap ukuran
# untuk mencari nilai rata rata menggunakan mean dengan ditambahkan tanda ()
# masukan variabel.groupby(["data yang ingin dikelompokan"]).mean()
# terdapat tanda titik untuk pemisah variabel dengan groupby
# diperhatikan selalu tanda "" di setiap data yang yang dipanggil
bunga_iris.groupby(["flower name"]).mean()
# sehingga hasil yang muncul adalah nilai rata rata di setiap ukuran pada setiap nama bunga 
```

```python
# setelah dikelompokan, pada contoh ini mengurutkan dari data terendah ke tertinggi
# untuk mengurutkannya menggunakan sort_values

# masukan variabel.groupby(["nama data yang ingin dikelompokan"]).mean().sort_values("data yang ingin diurutkan", ascending = True)
# diperhatikan tanda titik sebagai pemisah diantara fungsi atau variabel atau paramter
# diperhatikan selalu tanda "" di setiap data yang yang dipanggil
bunga_iris.groupby(["flower name"]).mean().sort_values('sepal width (cm)',ascending = True)
# sehingga hasil yang muncul adalah mengurutkan data sepal width melalui nilai rata ratanya
```

```python
# pada contoh ini mencari jumlah nilai pada masing masing ukuran berdasarkan nama bunganya
# untuk mencari nilai jumlah menggunakan nilai sum

# masukan variabelnya.groupby(["nama data yang dikelompokan"]).sum()
# di akhir sum selalu berisi tanda ()
# terdapat tanda titik untuk pemisah variabel dengan groupby
# diperhatikan selalu tanda "" disetiap memanggil nama data
bunga_iris.groupby(["flower name"]).sum()
# sehingga hasil yang muncul adalah jumlah nilai disetiap ukuran berdasarakan nama bunganya
```

```python
# pada contoh ini mengelompokan berdasarkan nama bunga untuk mencari jumlah di masing masing nama bunga
# untuk menghitung banyaknya data digunakan count
# masukan variabelnya.groupby(["nama data yang ingin dikelompokan"]).count()
# di setiap akhir count berisi tanda ()
bunga_iris.groupby(["flower name"]).count()
# sehingga hasil yang muncul adalah banyaknya data disetiap ukuran pada setiap nama bunga
```

### 5.22.4 Properti Pembuatan Histogram dengan Matplotlib

#### CONTOH 1

```python
# membuat histogram dengan bentuk dan warna sesuai keinginan

# masukan data dari flower name 
# nama disetiap flower name berisi tanda ""
name_flower = ['setosa','versicolor','virginica']
# masukan rata rata sepal length disetiap flower name 
# karena angka jadi tidak berisi tanda ""
sepal_length = [5.006,5.936,6.588]

# plt.figure bertujuan memberikan ukuran histogram yang diinginkan
plt.figure(figsize=(8,3))
# plt.title bertujuan untuk memberikan judul pada histogram
plt.title(("Grafik rata rata sepal length"))
# nah disini memasukan sumbu x dan y nya, serta memasukan warna nya
histogram = plt.bar(name_flower,sepal_length, color = "purple")

# disini memberikan simbol pada grafik di setiap nama bunga
# beda nama bunga beda simbol

# pada histogram [0] atau urutan pertama dengan memberi simbol titik
histogram[0].set_hatch(".")
# pada histogram [1] atau urutan kedua dengan memberi simbol x
histogram[1].set_hatch("x") 
# pada histogram [2] atau urutan ketiga dengan memberi simbol /
histogram[2].set_hatch("/")

# plt.show digunakan untuk menampilkan hasil histogram yang telah dibuat
plt.show()
```

```python
# data grafik dari 3 nama bunga beserta masing masing rata rata ukuran

# masukan data dari flower name 
# nama disetiap flower name berisi tanda ""
name_flower = ('setosa','versicolor','virginica')
# masukan data dari semua ukuran
ukuran_mean = {
    'sepal length': (5.006,5.936,6.588),
    'sepal width': (3.428,2.770,2.974),
    'petal length': (1.462,4.260,5.552),
    'petal width': (0.246,1.326,2.026)
}

x = np.arange(len(name_flower))
width = 0.25 
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in ukuran_mean.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=4)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('cm')
ax.set_title('Ukuran Rata Rata')
ax.set_xticks(x + width, name_flower)
ax.set_ylim(0, 10)

plt.show()
```
